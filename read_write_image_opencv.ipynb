# Read and display Image
this section of the file is for reading and displaying the downloaded inage from directory.

```python
# import open cv for reading and displaying images
import cv2
# import pandas for data manupulation
import pandas as pd
# import pyplot from matplotlib for ploting image.
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# read image using imread function. 1st argument is image file with path and 2nd argument
# IMREAD_GRAYSCALE is for reading the image as graysale.
img = cv2.imread('Opencv_logo.jpg',cv2.IMREAD_GRAYSCALE)
# imshow is for displaying image. 1st argument is title of image and 2nd argument is readen image file name.
cv2.imshow('image',img)
# waitkey function,as name suggest will wait for specified time period before returning. 0 means it will wait forever.
cv2.waitKey(0)
# at last kill all generated window.
cv2.destroyAllWindows()
```

Image read with the help of open cv can also be dispalyed by matplotlib


```python
img = cv2.imread('Opencv_logo.jpg',cv2.IMREAD_GRAYSCALE)
# imshow is used to display an imag eon axes.
plt.imshow(img,cmap='gray',interpolation='bicubic')
# disabling the tick value on x and y axes
plt.xticks([])
plt.yticks([])
# ploting the image
plt.show()
```

writing image file back to disk

```python
cv2.imwrite ('grayOpen_cvLogo.png',img)
```
